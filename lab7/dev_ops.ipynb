```python
from fabric import Connection

c = Connection('lab7')
```

```python
c.sudo('apt-get update')
c.sudo('apt-get install -y nginx')
c.sudo('apt-get install -y python3-venv')
c.sudo('apt-get install -y unzip ')
```

```python
c.run("mkdir -p ~/mysite")
with c.cd('~/mysite'):
    c.run("python3 -m venv myvenv")
    with c.prefix('source ~/mysite/myvenv/bin/activate'):
        c.run('pip install django')
        c.run('django-admin startproject lab')
        with c.cd('~/mysite/lab'):
            c.run('python3 manage.py startapp polls')
        
        c.put("./app.zip", "/home/ubuntu/mysite")
        c.run('unzip -o ~/mysite/app.zip -d ~/mysite/lab')

        
```

```python
c.put("./nginx.cfg", "/home/ubuntu/mysite")
c.sudo("mv nginx.cfg /etc/nginx/sites-enabled/default")
c.sudo("service nginx restart")
```

```python
with c.cd('~/mysite'):
    with c.prefix('source ~/mysite/myvenv/bin/activate'):
        with c.cd('~/mysite/lab'):
            c.run('python3 manage.py runserver 8000 &')
```
